# Analisis Model Regresi Linier

Kita akan mempersiapkan data untuk pelatihan model regresi linier. Langkah-langkahnya adalah sebagai berikut:

1. Split Data: Memisahkan data menjadi set pelatihan dan pengujian dengan rasio 70:30.
2. Pelatihan Model: Menggunakan model regresi linier pada set pelatihan.
3. Evaluasi Model: Menggunakan metrik Mean Absolute Error (MAE), Mean Squared Error (MSE), Mean Absolute Percentage Error (MAPE), dan koefisien determinasi (R²) untuk mengevaluasi model pada set pengujian.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# menentukan fitur dan target
X = data[['enginesize', 'cylinders', 'fuelconsumption']]
y = data['co2emissions']

# split data training dan testing (rasio 70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# training model Regres Linier
model = LinearRegression()
model.fit(X_train, y_train)

# prediksi data tes
y_pred = model.predict(X_test)

# membuat evaluasi model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
r2 = r2_score(y_test, y_pred)

metrics = {
    "Mean Absolute Error (MAE)": mae,
    "Mean Squared Error (MSE)": mse,
    "Mean Absolute Percentage Error (MAPE)": mape,
    "R-squared (R2)": r2
}

metrics
```

Berikut adalah metrik kinerjanya:
- Mean Absolute Error (MAE): 16.54
- Mean Squared Error (MSE): 503.60
- Mean Absolute Percentage Error (MAPE): 6.76%
- R-squared (R2): 0.875

Interpretasi Metrik
- MAE dan MSE yang relatif rendah menunjukkan bahwa kesalahan prediksi model rata-rata tidak terlalu besar.
- MAPE sebesar 6.76% menunjukkan bahwa model, rata-rata, memiliki kesalahan sekitar 6.76% dari nilai sebenarnya.
- R² mendekati 1, yang menunjukkan bahwa model cukup baik dalam menjelaskan variabilitas data.

```python
# visualisasi data asli vs prediksi
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Data Asli')
plt.ylabel('Prediksi')
plt.title('Data Asli vs Prediksi Emisi CO2')
plt.show()
```

Visualisasi di atas menampilkan nilai emisi CO2 yang sebenarnya (sumbu x) dibandingkan dengan nilai yang diprediksi oleh model regresi linier (sumbu y). Garis putus-putus mewakili lokasi di mana nilai prediksi sempurna akan jatuh.
